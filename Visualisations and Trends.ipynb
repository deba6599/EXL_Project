## Top 5 worst affected states

```python
grouped = x1['confirmed_cases'].groupby(x1['stateFIPS']).max()
```

```python
grouped = grouped.astype('int32')
grouped.nlargest(5)
```

```python
grouped.plot(kind="bar",figsize=(12,10),title="Worst affected states")
```

```python
x2 = x1[x1["stateFIPS"]==6]
x2
```

```python
x = x1[["stateFIPS","date","confirmed_cases","C_TOT_FEMALE"]]
x
```

```python
x2["date"]
```

```python
grouped1 = x2['confirmed_cases'].groupby(x2['date']).mean()
grouped1 = pd.concat([grouped1[15:],grouped1[:15]],axis=0)           #22-01-2020 start
grouped1.plot(kind="line",figsize=(20,15),title="Top 5 worst affected states trend")
```

```python
lst = [6,4,17,48,12]

fig, ax = plt.subplots()

for i in range(len(lst)):
    y = x[x["stateFIPS"]==i]
    gp1 = y['confirmed_cases'].groupby(y['date']).sum()
    gp1 = pd.concat([gp1[15:],gp1[:15]],axis=0)
    ax = gp1.plot(ax=ax,kind='line',label=lst[i],figsize=(20,10))
    


#fig, ax = plt.subplots()

#for key, grp in df.groupby(['color']):
#    ax = grp.plot(ax=ax, kind='line', x='x', y='y', c=key, label=key)

plt.legend(loc='best')
plt.show()
```

## Public data exploration

The following dataset is obtained from the website of Centers for Disease Control and Prevention(CDC) 

```python
a = pd.read_csv("Public_Data\Provisional_COVID-19_Death_Counts_by_Sex__Age__and_State.csv")
a
```

```python
a.isnull().sum()
```

```python
a = a.dropna(subset=["COVID-19 Deaths"])
```

```python
sns.relplot(x="Sex",y="COVID-19 Deaths",kind="line",data=a)
```

```python
b = a['COVID-19 Deaths'].groupby(a['Sex']).sum()
b.plot(kind="bar",figsize=(7,5),title="COVID-19 Deaths vs Sex")
```

```python
c = a['COVID-19 Deaths'].groupby(a['Age Group']).sum()
c.plot(kind="bar",figsize=(7,5),title="COVID-19 Deaths vs Age")
```

```python
p = pd.read_csv("Public_Data\Conditions_contributing_to_deaths_involving_coronavirus_disease_2019__COVID-19___by_age_group_and_state__United_States..csv")
p
```

```python
p.isnull().sum()
```

```python
p = p.dropna(subset=["COVID-19 Deaths"])
```

```python
q = p['COVID-19 Deaths'].groupby(p['Condition']).sum()
q.plot(kind="bar",figsize=(7,5),title="COVID-19 Deaths vs Condition")
```

```python
r = p['COVID-19 Deaths'].groupby(p['Condition Group']).sum()
r.plot(kind="bar",figsize=(7,5),title="COVID-19 Deaths vs Health Condition")
```

## Combining attributes

```python
corr_matrix = x1.corr()
```

```python
cm = corr_matrix["confirmed_cases"].sort_values(ascending=False)
cm
```

```python
x1["total_patients_confirmed_per_test_rate"] = x1["total_adult_patients_hospitalized_confirmed_covid_SD"]/x1["test_rate"]
```

```python
x1["total_patients_confirmed_hosp_per_tot_pop"] = x1["total_adult_patients_hospitalized_confirmed_covid_SD"]/x1["C_TOT_POP"]
```

```python
x1["total_patients_confirmed_hosp_per_beds"] = x1["total_adult_patients_hospitalized_confirmed_covid_SD"]/x1["inpatient_beds_SD"]
```

```python
x1["Employed_per_workplaces"] = x1["google_mobility_workplaces"]/x1["C_M_Employed_corr"]
```

```python
x1["test_count_per_labour"] = x1["test_count"]/x1["C_M_Labour_Force_corr"]
```

```python
corr_matrix_new = x1.corr()
cm_new = corr_matrix_new["confirmed_cases"].sort_values(ascending=False)
cm_new
```
