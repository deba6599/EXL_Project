```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x1 = pd.read_csv("PastSub\subrf1.csv")
x1
```

```python
x2 = pd.read_csv("MeanImputedNotScaled_dated.csv")
x2
```

```python
x2.info()
```

```python
corr_matrix = x2.corr() 
```

```python
cm = corr_matrix["confirmed_cases"].sort_values(ascending=False)
cm
```

```python
cm.index[2]
```

```python
cml = []
cl = []
for i in range(62):
    if cm[i] > 0.1 or cm[i]< 0.02:
        cml.append([cm.index[i],cm[i]])
        cl.append(cm.index[i])
cml,len(cl)
```

```python
cl.remove('Electricity_Sales_ Customers_Count_Industrial')
```

```python
val = pd.read_csv("EQ-2021-Validation-Data-Sample.csv")
val["confirmed_cases"] = x1["Confirmed_cases"]
val = val[cl]
val
```

```python
from sklearn.impute import SimpleImputer
mean_imputer = SimpleImputer(strategy="mean")

val_impt = mean_imputer.fit_transform(val)
val_impt
```

```python
Imp_val = pd.DataFrame(val_impt,columns=val.columns)
Imp_val
```

```python
Imp_val.info()
```

```python
x2= x2[cl]
x2
```

```python
Tab_df = pd.concat([x2,Imp_val],axis=0)
Tab_df
```

```python
Tab_df.to_csv("Tab_vis_time_2.csv",index=False)
```
