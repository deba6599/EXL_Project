## Importing libraries

```python
import numpy as np
import pandas as pd
```

## Reading files

```python
x = pd.read_csv("ScaledFeatures.csv")
y = pd.read_csv("labels.csv")
```

```python
x.head()
```

```python
x["label"] = y
```

```python
x.isnull().sum()
```

## Getting the correlation matrix

```python
corr_matrix = x.corr()
```

```python
corr_matrix["label"].sort_values(ascending=False)
```

```python
cm = corr_matrix["label"].sort_values(ascending=False)
```

```python
cm.index[2],cm[1]
```

## Getting Features with high positive or negative correlation

```python
imp = []
for i in range(63):
    if cm[i]>0.1 or cm[i]<0.01:
        imp.append((cm.index[i]))
```

```python
imp.remove("label")
```

```python
len(imp),imp
```

```python
x_imp = x[imp]
x_imp
```
