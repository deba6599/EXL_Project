```python
imf = pd.read_csv("ImpVal40.csv")
```

```python
val1 = val1[imf.columns.to_list()]
```

```python
val1.head()
```

```python
imf.head()
```

```python
val1["date"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
housing_cat = val1["date"]
housing_cat_encoded = encoder.fit_transform(housing_cat)
housing_cat_encoded
```

```python
val1 = val1.drop("date",axis=1)
```

```python
val1["date"] = housing_cat_encoded
```

```python
#val1.to_csv("ValDateLabelEncoded.csv",index=False)
```

```python
from sklearn.impute import SimpleImputer
```

```python
mean_imputer = SimpleImputer(strategy="mean")

ex_impt = mean_imputer.fit_transform(val1)
ex_impt
```

```python
Imp_val = pd.DataFrame(ex_impt,columns=val1.columns)
Imp_val
```

```python
Imp_val.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() 
standardized_features = scaler.fit_transform(Imp_val)
standardized_features
```

```python
scal_val = pd.DataFrame(standardized_features,columns=Imp_val.columns)
scal_val
```
