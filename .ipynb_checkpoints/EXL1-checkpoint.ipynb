```python
from statistics import mean
from tqdm import tqdm
def HandleDots(col):
    
    c = []
    l = ex1_cat[col].shape[0]
    for j in tqdm(range(l),desc="Progress"):
        if ex1_cat[col][j]=='.':
            pass
        else:
            c.append(int(ex1_cat[col][j]))
    
    m = mean(c)
    res = ex1_cat[col].replace(to_replace=".",value=m)
    res = res.astype("int32")
    return res    
```

```python
from statistics import mean            # all at once
from tqdm import tqdm
def HandleDots(lst):
    for i in tqdm(range(len(lst)),desc="List"):
        
    
        col = lst[i]
        c = []
        l = ex1_cat[col].shape[0]
        for j in tqdm(range(l),desc="Progress"):
            if ex1_cat[col][j]=='.':
                pass
            else:
                c.append(float(ex1_cat[col][j]))

        m = mean(c)
        res = dt
        res = ex1_cat[col].replace(to_replace=".",value=m)
        res = res.astype("float64")
        dt[col]=res
        
    return dt
```
